# body_mass_g / flipper_length_mm 관계 분석

```python
#1. 상관계수 검정 (Pearson's Correlation Test)
from scipy.stats import pearsonr

# Pearson's 상관계수와 p-value 계산
clean_df = df[['flipper_length_mm', 'body_mass_g']].dropna()
corr, p_value = pearsonr(clean_df['flipper_length_mm'], clean_df['body_mass_g'])
print(f"Pearson correlation coefficient: {corr}")
print(f"P-value: {p_value}")

```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# 산점도 및 회귀선 시각화
plt.figure(figsize=(8, 6))
sns.regplot(x='flipper_length_mm', y='body_mass_g', data=df, scatter_kws={'s': 50, 'alpha': 0.5}, line_kws={'color': 'red'})
plt.title('Scatter Plot with Regression Line')
plt.xlabel('Flipper Length (mm)')
plt.ylabel('Body Mass (g)')
plt.show()
```

## 성별에 따른 상관관계

```python
from scipy.stats import pearsonr

# 남성 펭귄에 대한 상관계수 계산
male_corr, male_p_value = pearsonr(male_df_copy['flipper_length_mm'], male_df_copy['body_mass_g'])
print(f"Male Pearson correlation coefficient: {male_corr}")
print(f"Male P-value: {male_p_value}")

# 여성 펭귄에 대한 상관계수 계산
female_corr, female_p_value = pearsonr(female_df_copy['flipper_length_mm'], female_df_copy['body_mass_g'])
print(f"Female Pearson correlation coefficient: {female_corr}")
print(f"Female P-value: {female_p_value}")
```

## 종에 따른 상관관계

```python
from scipy.stats import pearsonr

# Adelie 종에 대한 상관계수 계산
adelie_corr, adelie_p_value = pearsonr(Adelie_df['flipper_length_mm'], Adelie_df['body_mass_g'])
print(f"Adelie Pearson correlation coefficient: {adelie_corr}")
print(f"Adelie P-value: {adelie_p_value}")

# Chinstrap 종에 대한 상관계수 계산
chinstrap_corr, chinstrap_p_value = pearsonr(Chinstrap_df['flipper_length_mm'], Chinstrap_df['body_mass_g'])
print(f"Chinstrap Pearson correlation coefficient: {chinstrap_corr}")
print(f"Chinstrap P-value: {chinstrap_p_value}")

# Gentoo 종에 대한 상관계수 계산
gentoo_corr, gentoo_p_value = pearsonr(Gentoo_df['flipper_length_mm'], Gentoo_df['body_mass_g'])
print(f"Gentoo Pearson correlation coefficient: {gentoo_corr}")
print(f"Gentoo P-value: {gentoo_p_value}")
```

# 결과
- 각 종에서 flipper_length_mm와 body_mass_g는 모두 양의 상관관계를 보이며, 그 강도는 Gentoo > Chinstrap > Adelie 순으로 강하게 나타남
- Gentoo 종은 flipper_length_mm와 body_mass_g 사이에 강한 양의 상관관계가 있으며, 통계적으로 매우 유의미함
- Gentoo 종의 펭귄들 사이에서 flipper_length_mm (날개 길이)와 body_mass_g (체중) 간의 관계가 상당히 강하게 나타난다는 의미
- 결론: 날개 길수록 체중 증가하며, Gentoo 종의 펭귄은 특히 그 경향이 두드러짐

```python
# k-means 클러스터링

from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler

# 데이터 스케일링 (K-means는 데이터의 스케일에 민감)
scaler = StandardScaler()
scaled_data = scaler.fit_transform(Gentoo_df[['flipper_length_mm', 'body_mass_g']])

# KMeans 군집화 (3개의 군집으로 가정)
kmeans = KMeans(n_clusters=3, random_state=0)
Gentoo_df['cluster'] = kmeans.fit_predict(scaled_data)

# 군집 결과 시각화
plt.scatter(Gentoo_df['flipper_length_mm'], Gentoo_df['body_mass_g'], c=Gentoo_df['cluster'], cmap='viridis')
plt.xlabel('Flipper Length (mm)')
plt.ylabel('Body Mass (g)')
plt.title('K-means Clustering: Gentoo Penguins')
plt.colorbar(label='Cluster')
plt.show()
```
