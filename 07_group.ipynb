```python
import seaborn as sns
import pandas as pd

df = sns.load_dataset('titanic')
```

```python
df.head()
```

```python
# 성별 기준으로 생존률의 평균 알아보기
```

```python
df_copy = df.copy()
```

### groupby()

```python
df_copy.groupby('sex').mean(numeric_only=True)
```

```python
df_copy.groupby(['sex', 'pclass']).mean(numeric_only=True)
```

```python
df_copy.groupby(['sex', 'pclass'])[['survived']].mean(numeric_only=True)
```

### reset_index()

```python
df_temp = df_copy.groupby(['sex', 'pclass'])[['survived']].mean(numeric_only=True)
df_temp
```

```python
df_temp.reset_index()
```

```python
df_copy.groupby(['sex', 'pclass'])[['survived', 'age']].max(numeric_only=True)
```

```python
df_copy.groupby(['sex', 'pclass'])[['survived', 'age']].agg('mean', 'sum')
```

```python
df_copy.groupby(['sex', 'pclass'])[['survived', 'age']].agg('mean', 'sum').reset_index()
```

### pivot_table

```python
# 내가 가진 데이터의 축을 바꾸거나 세로와 가로를 바꿈
```

```python
df_copy.pivot_table(index='who', values='survived')
```

```python
df_copy.pivot_table(index=['who', 'pclass'], values=['survived', 'age'])
```
