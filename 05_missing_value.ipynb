```python
# 실습
cond1 = df_copy['sex'] == 'female'
woman_mean_age = df_copy[cond1]['age'].mean()
df_copy['age'] = df_copy['age'].fillna(woman_mean_age)

cond2 = df_copy['sex'] == 'male'
man_mean_age = df_copy[cond2]['age'].mean()
df_copy['age'] = df_copy['age'].fillna(man_mean_age)

df_copy['age'].mean()
```

```python
# 해설

df_copy = df.copy()
df_copy.head()
```

```python
cond1 = df_copy['sex'] == 'female'
woman_mean_age = df_copy.loc[cond1, 'age'].mean()
df_copy.loc[cond1, 'age'] = df_copy.loc[cond1, 'age'].fillna(woman_mean_age)

cond2 = df_copy['sex'] == 'male'
man_mean_age = df_copy.loc[cond2, 'age'].mean()
df_copy.loc[cond2, 'age'] = df_copy.loc[cond2, 'age'].fillna(man_mean_age)

df_copy['age'].mean()
```

```python
df_copy[['age', 'sex']].to_csv('data/age_change_final.csv')
```
