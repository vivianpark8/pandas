## 시계열 정보 접근 연습하기 - 자전거 데이터

```python
df = pd.read_csv('data/seoul_bicycle.csv')
df.head()
```

```python
df.info()
```

```python
df_copy = df.copy()
```

```python
df_copy['대여일자'] = pd.to_datetime(df_copy['대여일자'], format='%b-%d-%Y')
df_copy['대여일자'].info()
# 내가 원하는 정보를 시계열로 변환하는 함수
```

```python
cond = df_copy['대여일자'].dt.dayofweek == 2
df_copy.loc[cond]
```

```python
df_copy.head()
```

```python
df_copy['운동량'].info()
```

```python
df_copy.loc[2344] # 결측치가 아니라 '\N'이라는 object 데이터가 있음
```

```python
df_copy['운동량'] = pd.to_numeric(df_copy['운동량'], errors='coerce')
```

```python
df_copy['운동량'].info()
```

### cut()

```python
bins = [0, 100, 200, 300, df_copy['운동량'].max()]
```

```python
bins
```

```python
pd.cut(df_copy['운동량'], bins) # (컬럼, 나누는 구간) -> 어느 범주에 있는지
```

```python
pd.cut(df_copy['운동량'], bins=10)
pd.cut(df_copy['운동량'], bins=10).value_counts()
```

```python
pd.qcut(df_copy['운동량'], q=10)
pd.qcut(df_copy['운동량'], q=10).value_counts()
```

```python
qcut_bins = [0, 0.2, 0.8, 1] # 직접 비율을 설정함
```

```python
df_copy['운동량_한글'] = pd.qcut(df_copy['운동량'], qcut_bins, labels=['적음', '보통', '많음'])
```
